#### 2. Display all the image links from Walt Disney wikipedia page.
Hint: use `.get()` to access information inside tags. Check out the documentation.

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/Walt_Disney'
````

````python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Walt_Disney"

headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)

#  Parse HTML properly
soup = BeautifulSoup(response.text, "html.parser")

#  Now find image tags
images = soup.find_all("img")

image_links = []

for img in images:
    src = img.get("src")

    if src:
        if src.startswith("//"):
            src = "https:" + src
        image_links.append(src)

print(image_links)


````

#### 2.1. List all language names and number of related articles in the order they appear in wikipedia.org.

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'

headers = {"User-Agent": "Mozilla/5.0"}

lang = requests.get(url, headers=headers)

print(lang.status_code)
````

````python
soup = BeautifulSoup(lang.content)
language = soup.find_all("a")

language
````

````python
languages = soup.find_all("a", class_="link-box")

results = []

for lang in languages:
    name = lang.find("strong").get_text(strip=True)
    articles = lang.find("small").get_text(strip=True)

    results.append((name, articles))

print(results)
````

#### 2.2. Display the top 10 languages by number of native speakers stored in a pandas dataframe.
Hint: After finding the correct table you want to analyse, you can use a nested **for** loop to find the elements row by row (check out the 'td' and 'tr' tags). <br>An easier way to do it is using pd.read_html(), check out documentation [here](https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.read_html.html).

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'

headers = {"User-Agent": "Mozilla/5.0"}

table = requests.get(url, headers=headers)

print(table.status_code)

````

````python
soup = BeautifulSoup(table.content)
table = soup.find("table")

table
````

````python
table = soup.find("table", {"class": "wikitable"})

rows = table.find_all("tr")

data = []

for row in rows[1:11]:   # Skip header, take top 10
    cols = row.find_all("td")
    if cols:
        language = cols[1].get_text(strip=True)
        speakers = cols[2].get_text(strip=True)
        data.append([language, speakers])

df = pd.DataFrame(data, columns=["Language", "Native Speakers"])
df
````

#### 3. Display Metacritic top 24 Best TV Shows of all time (TV Show name, initial release date, metascore rating, film rating system and description) as a pandas dataframe.
Hint: If you hover over the title of the movie, you should see the director's name. Can you find where it's stored in the html?

````python
# This is the url you will scrape in this exercise
url = 'https://www.metacritic.com/browse/tv/'


headers = {"User-Agent": "Mozilla/5.0"}

show = requests.get(url, headers=headers)
print(show.status_code)
````

````python
soup = BeautifulSoup(show.content)
shows = soup.find_all("div")[:24]

shows
````

````python
shows = soup.find_all("div", class_="c-finderProductCard")[:24]

data = []

for show in shows:

    # Title
    title_tag = show.find("h3")
    title = title_tag.get_text(strip=True) if title_tag else None

    # Release Date
    release_tag = show.find("span", class_="c-finderProductCard_meta")
    release_date = release_tag.get_text(strip=True) if release_tag else None

    # Metascore
    score_tag = show.find("div", class_="c-siteReviewScore")
    metascore = score_tag.get_text(strip=True) if score_tag else None

    # Rating system (TV-MA etc.)
    rating_tag = show.find("span", class_="c-finderProductCard_rating")
    rating = rating_tag.get_text(strip=True) if rating_tag else None

    # Description
    description_tag = show.find("div", class_="c-finderProductCard_description")
    description = description_tag.get_text(strip=True) if description_tag else None

    data.append([
        title,
        release_date,
        metascore,
        rating,
        description
    ])

df = pd.DataFrame(data, columns=[
    "TV Show",
    "Initial Release Date",
    "Metascore",
    "Rating",
    "Description"
])

df
````

#### 3.1. Find the image source link and the TV show link. After you're able to retrieve, add them to your initial dataframe

````python
data = []

for show in shows:

    # Title
    title_tag = show.find("h3")
    title = title_tag.get_text(strip=True) if title_tag else None

    # Release Date
    release_tag = show.find("span", class_="c-finderProductCard_meta")
    release_date = release_tag.get_text(strip=True) if release_tag else None

    # Metascore
    score_tag = show.find("div", class_="c-siteReviewScore")
    metascore = score_tag.get_text(strip=True) if score_tag else None

    # Rating
    rating_tag = show.find("span", class_="c-finderProductCard_rating")
    rating = rating_tag.get_text(strip=True) if rating_tag else None

    # Description
    description_tag = show.find("div", class_="c-finderProductCard_description")
    description = description_tag.get_text(strip=True) if description_tag else None

    # 🔹 TV Show Link
    link_tag = show.find("a")
    show_link = None
    if link_tag and link_tag.get("href"):
        show_link = "https://www.metacritic.com" + link_tag.get("href")

    # 🔹 Image Source
    img_tag = show.find("img")
    image_src = img_tag.get("src") if img_tag else None

    data.append([
        title,
        release_date,
        metascore,
        rating,
        description,
        show_link,
        image_src
    ])

df = pd.DataFrame(data, columns=[
    "TV Show",
    "Initial Release Date",
    "Metascore",
    "Rating",
    "Description",
    "TV Show Link",
    "Image Source"
])

df
````
